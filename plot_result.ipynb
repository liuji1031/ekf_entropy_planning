```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data_random_1 = np.loadtxt("loss_05_06_2024-14_17_16_random_policy.txt")
data_random_2 = np.loadtxt("loss_05_06_2024-14_17_43_random_policy.txt")
data_random = np.concatenate([data_random_1, data_random_2],axis=0)
print(data_random.shape)

data_plan_1 = np.loadtxt(
    "loss_05_06_2024-13_38_55_entropy_planning.txt").flatten()
data_plan_2 = np.loadtxt(
    "loss_05_06_2024-13_41_26_entropy_planning.txt").flatten()
```

```python
# plot mean and std
m = data_random.mean(axis=0)
s = data_random.std(axis=0)
x = np.arange(0,m.shape[0])
plt.plot(x,m, color='r',label='random')
plt.fill_between(x, m-s, m+s,
    alpha=0.2, facecolor='r')

# plot entropy planning runs
plt.plot(x,data_plan_1,color='b',label="planning 1")
plt.plot(x,data_plan_2,color='b',linestyle="-.",label="planning 2")

ax = plt.gca()
ax.spines[['right', 'top']].set_visible(False)

ax.set_xlabel("Time step")
ax.set_ylabel("Max eig. value")

plt.legend()
plt.savefig("eigenvalue_change.png",dpi=300)
plt.show()
```
